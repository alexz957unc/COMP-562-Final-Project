```python
rows = 4
columns = 6
row_idx = 0
col_idx = 0

fig, axes = plt.subplots(rows, columns, figsize=(20, 20))#Boxplot


for col in df.columns:
    try:
        sns.boxplot(x = col, data=df, ax=axes[row_idx, col_idx])
        if col_idx + 1 == columns:
            col_idx = 0
            row_idx += 1
        else:
            col_idx += 1
    except TypeError:
        pass
```

```python
corr = df.corr()
corr
```

```python
#Colinearity plot
plt.figure(figsize=(25,25))
sns.heatmap(corr, cmap='RdBu_r', annot=True, vmax=1, vmin=-1)
```
