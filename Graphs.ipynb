```python
import matplotlib.pyplot as plt
Models = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'KNN', 'SVM', 'XGBoost', 'AdaBoost']
Accuracy = [0.7485, 0.73831	, 0.7509, 0.72558, 0.74843, 0.75416, 0.75536]
F1 = [0.74839, 0.73806, 0.75038, 0.72465, 0.74796, 0.75368, 0.75502]
Precision = [0.74901, 0.73936, 0.75323	, 0.72894, 0.75049, 0.75635, 0.75694]
Recall = [0.7485, 0.73831, 0.7509, 0.72558, 0.74843, 0.75416, 0.75536]
```

```python
sorted_data = sorted(zip(Accuracy, Models), reverse=False)
sorted_accuracy, sorted_models = zip(*sorted_data)

plt.barh(sorted_models, sorted_accuracy)
plt.xlabel('Accuracy')
plt.ylabel('Models')
plt.title('Accuracy of different models Bar Graph')
plt.show()
```
